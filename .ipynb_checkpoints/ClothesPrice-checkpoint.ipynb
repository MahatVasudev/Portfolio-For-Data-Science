```python
def brand_filter(val):
    val = val.strip()
    val = val.title()
    db = {"Reebok":1,"Under Armour":2,"Nike":3,"Adidas":4,"Puma":5,"New Balance":6}
    if val not in db.keys():
        raise ValueError("The Value Doesnt Exist, Try Again")
    else:
        return db[val]
```

```python
class efficient:
    def __init__(self):
        self.brand = {"Reebok":1,"Under Armour":2,"Nike":3,"Adidas":4,"Puma":5,"New Balance":6}
        self.category = {"Dress":1,"Jeans":2,"Shoes":3,"Sweater":4,"Jacket":5,"T-Shirt":6}
        self.color = {"White":1,"Black":2,"Red":3,"Green":4,"Yellow":5,"Blue":6}
        self.size = ({"S":2,"M":3,"L":4,"Xs":1,"Xl":4,"Xxl":5})
        self.materials = {'Cotton': 0, 'Denim': 1, 'Nylon': 2, 'Polyester': 3, 'Silk': 4, 'Wool': 5}
    def brand_filter(self,val):
        return self.clean_n_validate(val,self.brand)

    def category_filter(self,val):
        return self.clean_n_validate(val,self.category)

    def color_filter(self,val):
        return self.clean_n_validate(val,self.color)

    def size_filter(self,val):
        return self.clean_n_validate(val,self.size)

    def material_filter(self,val):
        return self.clean_n_validate(val,self.materials)
        
    def clean_n_validate(self,val,db):
        orig,val = val, val.strip().title()
        if val not in db.keys():
            raise ValueError(f"{orig} : Value Doesnt Exist, Try Again")
        else:
            return db[val],orig
```

```python
def main():
    e = efficient()
    print("-"*50)
    print("\tEnter Brand \n\tAvailable Brands\n\t1.Reebok, 2.Under Armour, 3.Nike, 4.Adidas, 5.Puma, 6.New Balance")
    Brand = e.brand_filter(input("->"))
    print("\tEnter Wearables Category \n\tAvailable Categories\n\t1.Dress, 2.Jeans, 3. Shoes, 4.Sweater, 5.Jackets, 6.T-shirt")
    category = e.category_filter(input("->"))
    print("\tEnter Color \n\tAvailable Colors\n\t1.White, 2.Black, 3.Red, 4.Green, 5.Yellow, 6.Blue")
    color = e.color_filter(input("->"))
    print("\tEnter Sizes \n\tAvailable Sizes\n\t1.XS, 2.S, 3.M, 4.L, 5.XL, 6.XXL")
    size = e.size_filter(input("->"))
    print("\tEnter Materials \n\tAvailable Materials\n\t1.Cotton, 2.Denim, 3.Nylon, 4.Polyester, 5.Wool")
    material = e.material_filter(input("->"))
    features = [[Brand[0],category[0],color[0],size[0],material[0]]]
    predict = model.predict(features)[0]
    return f"{[Brand[1],category[1],color[1],size[1],material[1]]} costs estm. {predict}$"
```
