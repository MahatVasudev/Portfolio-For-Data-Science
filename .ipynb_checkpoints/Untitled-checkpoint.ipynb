```python
import sys

from KMeans import KMeans
import numpy as np
```

```python
from sklearn.datasets import make_blobs
```

```python
from sklearn.datasets import load_iris
```

```python
import matplotlib.pyplot as plt
```

```python
X,y = make_blobs(n_samples = 500,n_features = 3,centers = 3,random_state = 23)
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
model = KMeans(k =3,rand_state=3123,max_iter=10)
model.fit(X)
ypred = model.predict_clusters()
clusters = model.getclusters()
print(clusters)
plt.scatter(X[:,0],X[:,1],c = ypred)
for i in clusters:
    center = clusters[i]['center']
    plt.scatter(center[0],center[1],marker='^',c='red')
```
